```python
import pandas as pd
from utils.translation_utils import * 
import numpy as np
```

```python
INCIDENTS_PATH_FR = 'data/Incidents métro_20260707.xls'
PLANNED_MILEAGE_PATH_FR = 'data/Kilométrage métro planifé_20260707.xls'

na_map = {
    "Non affecté": "unassigned",
    "#": "N/A"
}
```

```python
incidents_df = pd.read_csv(INCIDENTS_PATH_FR, sep=";").replace(na_map)
incidents_df = incidents_df.where(pd.notnull(incidents_df), None)
mileage_df = pd.read_csv(PLANNED_MILEAGE_PATH_FR, sep=",", dtype={'Tournée':str}).replace(na_map)
mileage_df = mileage_df.where(pd.notnull(mileage_df), None)
```

```python
incident_translations = SCHEMA['incidents']
mileage_translations = SCHEMA['mileage']
incidents_df_en = incidents_df.rename(columns={elem['fr_name']: elem['en_name'] for elem in incident_translations})
mileage_df_en = mileage_df.rename(columns={elem['fr_name']: elem['en_name'] for elem in mileage_translations})
```

```python
for col in incident_translations:
    if col['values']:
        name = col['en_name']
        incidents_df_en[name] = apply_map(incidents_df_en[name], col['values'])

incidents_df_en.to_csv("data/metro_incidents_en.csv", index=False)
```

```python
for col in mileage_translations:
    if col['values']:
        name = col['en_name']
        mileage_df_en[name] = apply_map(mileage_df_en[name], col['values'])

mileage_df_en.to_csv("data/planned_mileage_en.csv", index=False)
```
